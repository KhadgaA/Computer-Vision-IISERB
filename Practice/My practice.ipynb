```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY).astype(float)
img
```

```python
edge_x = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 3)
edge_y = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 3)
edge = np.sqrt(edge_x**2 + edge_y**2)

# cv2.imshow('wind',edge)
# cv2.imwrite('beach_convol.jpg', edge)
# edge
```

```python
plt.imshow(edge, cmap='gray')
```

```python
# kernal = np.array([[1,1,1],[1,1,1],[1,1,1]])
kernal = np.array([[0,0,0],
                   [0,0,1],
                   [0,0,0]])
# kernal = kernal/sum(kernal)
img_lpf = cv2.filter2D(img,-1,kernal)
# cv2.imwrite('lpf.jpg',img_lpf)
plt.imshow(img_lpf, cmap='gray')
```

```python
kernel = np.array([[0.0, -1.0, 0.0], 
                   [-1.0, 4.0, -1.0],
                   [0.0, -1.0, 0.0]])

kernel = kernel/(np.sum(kernel) if np.sum(kernel)!=0 else 1)

#filter the source image
img_hpf = cv2.filter2D(img,-1,kernel)

#save result image
cv2.imwrite('hpf.jpg',img_hpf)
```

```python
kernel = np.array([[-1.0, -1.0, -1.0], 
                   [-1.0, 8.0, -1.0],
                   [-1.0, -1.0, -1.0]])

kernel = kernel/(np.sum(kernel) if np.sum(kernel)!=0 else 1)

#filter the source image
img_hpf = cv2.filter2D(img,-1,kernel)

#save result image
cv2.imwrite('hpf2_c.jpg',img_hpf)
```

```python
imgx = img + 0.5*img_hpf
cv2.imwrite('sharp_c.jpg',imgx)
```
