```python
N = 501;
x2 = (N-1)/2;
x1 = -x2
x,y= np.mgrid[x1:x2+1,x1:x2+1]
x.shape
```

```python
r = np.sqrt(x**2 + y**2)
r.shape
```

```python
x
```

```python
import math 
km = 0.7*np.pi
rm = x2
w = rm/10
term1 = np.sin( (km * r**2) / (2 * rm) )
term2 = 0.5*np.tanh((rm - r)/w) + 0.5
g = term1 * term2
I = (g + 1)/2
```

#### Q7.1

```python
center = np.array(image.shape[0:2]) / 2
x = center[1] - 250/2
y = center[0] - 250/2

I_cropped = image[int(y):int(y+250), int(x):int(x+250)]
plt.figure(figsize = (10,10))
plt.imshow(I_cropped,cmap='gray')
```

#### Q7.2

```python
kernal = np.array([[1,1,1],[1,1,1],[1,1,1]])
kernal = kernal/sum(kernal)
img_lpf = cv2.filter2D(I_cropped,-1,kernal)
plt.figure(figsize = (10,10))
plt.imshow(img_lpf,cmap='gray')
# cv2.imwrite('lpf.jpg',img_lpf)
```

```python
blur_img=cv2.GaussianBlur(img_lpf,(5,5),0)
plt.figure(figsize = (10,10))
plt.imshow( blur_img,cmap = 'gray')
```
